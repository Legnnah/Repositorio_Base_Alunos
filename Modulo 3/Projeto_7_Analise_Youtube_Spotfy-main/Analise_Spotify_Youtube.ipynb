### 6. Quais os 5 artistas com mais streams no spotify?

```python
### Quais os 5 artistas com mais streams no spotify?
stream_por_artista = df.groupby("Artist")["Stream"].sum()
stream_por_artista.sort_values(ascending=False).head(5)
```
